```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import math

path = '../data/processed/'
filename = 'mcc_clean_churn.csv'

df = pd.read_csv(path+filename)

df.columns

df.head(5)
```

```python
#def cal_phi_for_I_D_vars(i_var, d_var, df=df):
i_var='Int_Plan'
d_var='Churned'
#df_mod=df

# KEEP Categorical vars = {Churned, Int_Plan, VM_Plan}

df_mod = df[[i_var, d_var,'Churned']]

df_mod.head()
```

```python
Phi_a = df_mod[(df_mod[i_var] == 0) & (df_mod['Churned'] == 0)].count()

Phi_b = df_mod[(df_mod['Int_Plan'] == 0) & (df_mod['Churned'] == 1)].count()

Phi_c = df_mod[(df_mod['Int_Plan'] == 1) & (df_mod['Churned'] == 0)].count()

Phi_d = df_mod[(df_mod['Int_Plan'] == 1) & (df_mod['Churned'] == 1)].count()

# Fisher 2x2 Contingency Table

print('Phi_a =', Phi_a[0])
print('Phi_b =', Phi_b[0])
print('Phi_c =', Phi_c[0])
print('Phi_d =', Phi_d[0])
```

```python
numerator = (Phi_a*Phi_d)-(Phi_b*Phi_c)
print(numerator)
denominator = ((Phi_a+Phi_d)*(Phi_c+Phi_d)*(Phi_a+Phi_c)*(Phi_b+Phi_d))**(0.5)
print(f'The numerator of Phi = {numerator}')
print(f'The denominator of Phi = {math.floor(denominator)}')
print(f'Phi Coefficient = {numerator/denominator:0.3f}')
```
